membaca data dari file dan menaruhnya pada variabel setiap kalimat dan katanya

```python
data = []
with io.open('data_sastra_jawa_full.txt', 'r') as file:
    for entry in file:
        entry = entry.strip()
        data.append(entry)
```

Translate bahasa Jawa ke bahasa English karena kebanyakan library hanya berbasis bahasa English

```python
print(data)
```

```python
from googletrans import Translator
import time

translator = Translator()
language = 'en'
sleep_time = 3

translated = []
data_jawa = []
for i in range(len(data)):
    try:
        if data[i]:  # cek apakah elemen tidak None atau kosong
            translation = translator.translate(data[i], dest=language)
            if translation and translation.text:  # cek apakah hasil terjemahan valid
                translated.append(translation.text)
                data_jawa.append(data[i])
            else:
                continue
        else:
            continue
    except Exception as e:
        print(f"Error in translating {data[i]}: {e}")
    time.sleep(sleep_time)

print(translated)
```

Text dalam bahasa English

```python
transtexts = translated
print(transtexts)
```

```python
import pandas as pd

array_kalimat = transtexts
array_nomor = []
for i in range(1, len(array_kalimat) + 1):
  array_nomor.append(i)
```

```python
# Menggabungkan kedua array menjadi DataFrame
df = pd.DataFrame({
    'Nomor': array_nomor,
    'Kalimat': array_kalimat
})

# Menyimpan DataFrame ke file Excel
df.to_excel('data_sastra_inggris_full.xlsx', index=False, engine='openpyxl')

print("Data telah disimpan ke file Excel.")
```
